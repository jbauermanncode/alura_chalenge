```python
import pandas as pd
```

```python
churn_de_clientes = pd.read_json('https://raw.githubusercontent.com/sthemonica/alura-voz/main/Dados/Telco-Customer-Churn.json')
```

```python
churn_de_clientes.head(5)
```

```python
conta = churn_de_clientes['account']
```

```python
conta.head(5)
```

```python
conta[0]
```

## tipo de contrato

```python
selecao = [i['Contract'] for i in conta]
```

```python
contrato = pd.DataFrame(selecao, columns = ['contrato'])
```

```python
contrato
```

```python
contrato.drop_duplicates()
```

```python
one_year = ['One year']
contrato_one_year = contrato['contrato'].isin(one_year)
contrato_one_year = contrato[contrato_one_year]
```

```python
contrato_one_year
```

```python
two_year = ['Two year']
contrato_two_year = contrato['contrato'].isin(two_year)
contrato_two_year = contrato[contrato_two_year]
```

```python
contrato_two_year
```

```python
month = ['Month-to-month']
contrato_month = contrato['contrato'].isin(month)
contrato_month = contrato[contrato_month]
```

```python
contrato_month
```

```python
soma_contrato = len(contrato_one_year) + len(contrato_two_year) + len(contrato_month)
```

```python
soma_contrato
```

```python
conta[0]
```
