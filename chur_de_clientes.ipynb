# Separar Clientes

```python
selecao_cliente = churn_de_clientes['customer']
```

```python
selecao_cliente
```

## Separar Clientes por genero

```python
genero = [d['gender'] for d in selecao_cliente]
```

```python
genero = pd.DataFrame(genero, columns = ['genero'])
```

```python
genero
```

```python
female = ['Female']
feminino = genero['genero'].isin(female)
```

```python
feminino = genero[feminino]
```

```python
male = ['Male']
masculino = genero['genero'].isin(male)
```

```python
masculino = genero[masculino]
```

```python
len(masculino) + len(feminino)
```

```python
print(f'Sexo Masculino: {len(masculino)}\nSexo Feminino: {len(feminino)}')
```

## Separar Clientes para saber se o cliente tem idade >= 65

```python
idoso = [d['SeniorCitizen'] for d in selecao_cliente]
```

```python
idoso = pd.DataFrame(idoso, columns = ['idoso'])
```

```python
idoso_yes = idoso['idoso'].isin([True])
```

```python
idoso_yes = idoso[idoso_yes]
```

```python
idoso_yes
```

```python
idoso_not = idoso['idoso'].isin([False])
```

```python
idoso_not = idoso[idoso_not]
```

```python
idoso_not
```

```python
len(idoso_yes) + len(idoso_not)
```

```python
print(f'Idade >= 65: {len(idoso_yes)}\nIdade < 65: {len(idoso_not)}')
```

## Separar se cliente tem ou não parceiro

```python
parceiro = [d['Partner'] for d in selecao_cliente]
```

```python
parceiro = pd.DataFrame(parceiro, columns = ['parceiro'])
```

```python
parceiro
```

```python
yes = ['Yes']
parceiro_yes = parceiro['parceiro'].isin(yes)
```

```python
parceiro_yes = parceiro[parceiro_yes]
```

```python
no = ['No']
parceiro_not = parceiro['parceiro'].isin(no)
```

```python
parceiro_not = parceiro[parceiro_not]
```

```python
len(parceiro_not) + len(parceiro_yes)
```

```python
print(f'Cliente tem parceiro: {len(parceiro_yes)}\nCliente não tem parceiro: {len(parceiro_not)}\n')
```

# Acrescentando colunas e excluindo colunas

```python
churn_de_clientes['tem parceiro'] = parceiro
```
