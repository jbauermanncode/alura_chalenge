## separar assinatura de um provedor internet

```python
selecao = [i['InternetService'] for i in internet]
```

```python
selecao
```

```python
servico_provedor = pd.DataFrame(selecao, columns = ['assinatura_provedor_internet'])
```

```python
servico_provedor.head(5)
```

```python
servico_provedor.drop_duplicates()
```

```python
dsl = ['DSL']
servico_provedor_DSL = servico_provedor['assinatura_provedor_internet'].isin(dsl)
servico_provedor_DSL = servico_provedor[servico_provedor_DSL]
```

```python
servico_provedor_DSL
```

```python
fiber_optic = ['Fiber optic']
servico_provedor_fiber_optic = servico_provedor['assinatura_provedor_internet'].isin(fiber_optic)
servico_provedor_fiber_optic = servico_provedor[servico_provedor_fiber_optic]
```

```python
servico_provedor_fiber_optic
```

```python
no = ['No']
servico_provedor_not = servico_provedor['assinatura_provedor_internet'].isin(no)
servico_provedor_not = servico_provedor[servico_provedor_not]
```

```python
servico_provedor_not
```

```python
soma_provedor = len(servico_provedor_DSL) + len(servico_provedor_fiber_optic) + len(servico_provedor_not)
```

```python
internet[0]
```
