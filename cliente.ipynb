```python
import pandas as pd
```

```python
churn_de_clientes = pd.read_json('https://raw.githubusercontent.com/sthemonica/alura-voz/main/Dados/Telco-Customer-Churn.json')
```

```python
churn_de_clientes.head(5)
```

```python
cliente = churn_de_clientes['customer']
```

```python
cliente[0]
```

```python
type(cliente)
```

## separar por genero

```python
cliente
```

```python
selecao = [i['gender'] for i in cliente]
```

```python
genero = pd.DataFrame(selecao, columns = ['genero'])
```

```python
genero.head(5)
```

```python
female = ['Female']
feminino = genero['genero'].isin(female)
feminino = genero[feminino]
```

```python
feminino
```

```python
male = ['Male']
masculino = genero['genero'].isin(male)
masculino = genero[masculino]
```

```python
masculino
```

```python
soma_genero = len(masculino) + len(feminino)
```

## separar clientes para saber se o cliente tem idade >= 65

```python
idoso = [d['SeniorCitizen'] for d in cliente]
```

```python
idoso = pd.DataFrame(idoso, columns = ['idoso'])
```

```python
idoso.head(5)
```

```python
idoso.drop_duplicates()
```

```python
idoso_maior_igual_65 = idoso['idoso'].isin([True])
idoso_maior_igual_65 = idoso[idoso_maior_igual_65]
```

```python
idoso_maior_igual_65
```

```python
idoso_menor_65 = idoso['idoso'].isin([False])
idoso_menor_65 = idoso[idoso_menor_65]
```

```python
idoso_menor_65
```

```python
soma_idoso = len(idoso_maior_igual_65) + len(idoso_menor_65)
```

## separar se cliente tem ou não parceiro

```python
parceiro = [d['Partner'] for d in cliente]
```

```python
parceiro = pd.DataFrame(parceiro, columns = ['parceiro'])
```

```python
parceiro
```

```python
parceiro.drop_duplicates()
```

```python
len(parceiro.notnull())
```

```python
yes = ['Yes']
parceiro_yes = parceiro['parceiro'].isin(yes)
parceiro_yes = parceiro[parceiro_yes]
```

```python
parceiro_yes
```

```python
no = ['No']
parceiro_not = parceiro['parceiro'].isin(no)
parceiro_not = parceiro[parceiro_not]
```

```python
parceiro_not
```

```python
soma_parceiro = len(parceiro_not) + len(parceiro_yes)
```

## separar se tem ou não tem dependentes

```python
dependentes = [d['Dependents'] for d in cliente]
```

```python
dependentes = pd.DataFrame(dependentes, columns = ['dependentes'])
```

```python
dependentes
```

```python
dependentes.drop_duplicates()
```

```python
len(dependentes.notnull())
```

```python
yes = ['Yes']
dependentes_yes = dependentes['dependentes'].isin(yes)
dependentes_yes = dependentes[dependentes_yes]
```

```python
dependentes_yes
```

```python
no = ['No']
dependentes_not = dependentes['dependentes'].isin(no)
dependentes_not = dependentes[dependentes_not]
```

```python
dependentes_not
```

```python
soma_dependentes = len(dependentes_yes) + len(dependentes_not)
```

## acrescentando e excluindo colunas

```python
churn_de_clientes_new = pd.read_csv('churn_de_clientes.csv')
```

```python
churn_de_clientes_new
```

```python
churn_de_clientes_new['genero'] = genero
```

```python
churn_de_clientes_new['idade_>=_65'] = idoso
```

```python
churn_de_clientes_new['tem_parceiro'] = parceiro
```

## criar DataFrame dados_pessoais

```python
clientes = pd.DataFrame(churn_de_clientes_new[['cliente_ID','Churn','genero','idade_>=_65','tem_parceiro','tem_dependentes']])
```

```python
clientes.to_csv('dados_pessoais.csv', index=False)
```
