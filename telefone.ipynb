```python
import pandas as pd
import numpy as np
```

```python
churn_de_clientes = pd.read_json('https://raw.githubusercontent.com/sthemonica/alura-voz/main/Dados/Telco-Customer-Churn.json')
```

```python
churn_de_clientes.head(5)
```

```python
selecao_phone = churn_de_clientes['phone']
```

```python
selecao_phone.head(5)
```

## Assinatura de serviço telefônico com uma linha

```python
servico_telefone = [p['PhoneService'] for p in selecao_phone]
```

```python
servico_telefone = pd.DataFrame(servico_telefone, columns = ['servico_telefone'])
```

```python
servico_telefone
```

```python
yes = ['Yes']
servico_telefone_yes = servico_telefone['servico_telefone'].isin(yes)
servico_telefone_yes =  servico_telefone[servico_telefone_yes]
```

```python
servico_telefone_yes
```

```python
no = ['No']
servico_telefone_not = servico_telefone['servico_telefone'].isin(no)
servico_telefone_not = servico_telefone[servico_telefone_not]
```

```python
servico_telefone_not
```

```python
soma_telefone = len(servico_telefone_yes) + len(servico_telefone_not)
```

## Assinatura de serviço telefônico com mais de uma linha

```python
selecao_multiplas_linhas = [p['MultipleLines'] for p in selecao_phone]
```

```python
multiplas_linhas = pd.DataFrame(selecao_multiplas_linhas, columns = ['servico_telefone_multiplas_linhas'])
```

```python
multiplas_linhas
```

```python
yes = ['Yes']
multiplas_linhas_yes = multiplas_linhas['servico_telefone_multiplas_linhas'].isin(yes)
multiplas_linhas_yes = multiplas_linhas[multiplas_linhas_yes]
```

```python
multiplas_linhas_yes
```

```python
no = ['No']
multiplas_linhas_not = multiplas_linhas['servico_telefone_multiplas_linhas'].isin(no)
multiplas_linhas_not = multiplas_linhas[multiplas_linhas_not]
```

```python
multiplas_linhas_not
```

```python
multiplas_linhas.drop_duplicates()
```

```python
selecao_diferente = ((multiplas_linhas['servico_telefone_multiplas_linhas'] != 'Yes') & (multiplas_linhas['servico_telefone_multiplas_linhas'] != 'No'))
```

```python
selecao_diferente
```

```python
multiplas_linhas_diferente = multiplas_linhas[selecao_diferente]
```

```python
multiplas_linhas_diferente
```

```python
soma_multiplas = len(multiplas_linhas_yes) + len(multiplas_linhas_not) + len(multiplas_linhas_diferente)
```

## acrescentando e excluindo colunas

```python
churn_de_clientes_new = pd.read_csv('churn_de_clientes.csv')
```

```python
churn_de_clientes_new.head(5)
```

```python
churn_de_clientes_new['servico_telefone'] = servico_telefone
```

```python
churn_de_clientes_new.head(5)
```

```python
churn_de_clientes_new['servico_telefone_multiplas_linhas'] = multiplas_linhas
```

```python
churn_de_clientes_new.head(5)
```

## DataFrame dados_telefonicos

```python
dados_telefonicos = pd.DataFrame(churn_de_clientes_new[['cliente_ID','servico_telefone', 'servico_telefone_multiplas_linhas']])
```

```python
dados_telefonicos.to_csv('dados_telefonicos.csv', index=False)
```
